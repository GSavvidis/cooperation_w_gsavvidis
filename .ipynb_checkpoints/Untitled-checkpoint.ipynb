```python
import sympy as sm
sm.init_printing(True)
```

```python
r,x2,x1,y2,y1 = sm.symbols('r x2 x1 y2 y1')
```

```python
V = (1/r)**12-(1/r)**6
```

```python
V
```

```python
V = V.subs(r,((x1-x2)**2+(y1-y2)**2)**1/2)
```

```python
V
```

```python
Vton = V.diff(x1)

```

```python
Vton
```

```python
Vton_x2 = V.diff(x2)
```

```python
Vton_x2
```

Παρατηρώ οτι η δυναμη που ασκει το χ1 στο χ2 ταυτίζεται με αυτή που ασκεί το x2 στο x1 μετα την εναλλαγή χ1<->χ2

```python
Vton_y2 = V.diff(y2)
```

```python
Vton_y2
```
